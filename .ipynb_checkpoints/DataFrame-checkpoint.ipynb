```python
import pandas as pd 
ls = [
    ["shubham",29,5],
    ["sanket", 28, 6], 
    ["rohan", 26,6]
]
print(ls)
```

```python
student_def = pd.DataFrame(ls , columns= ["name" , "age" , "package"])
print(student_def)
```

```python
student_dict = {
    "name" : ["shubham","sanket","rohan"],
    "age" : [28, 26 , 25], 
    "package" : [5,6,7]
}
student_def = pd.DataFrame(student_dict)
print(student_def)
```
