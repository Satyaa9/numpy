```python
import pandas as pd
students = pd.DataFrame(
    {
        'student_id' : [1,2,3,4], 
        'name' : ["shubham" , "sanket" , "rohan" , "rajat"], 
        'city' : ["pune" , "satara" , "pune" , "mumbai"]
    }
)
students
```

```python
marks = pd.DataFrame(
    {
        'student_id' : [1,2,3,5],
        'Maths' : [70,90,75,80],
        'Science' : [60,70,40,80]
    }
)
marks
```

```python
ExtraStudent = pd.DataFrame(
    {
        'student_id' : [6,7], 
        'name' : ["swapnil" , "ghansham"], 
        'city' : ["solapur" , "piliv"]
    }
)
ExtraStudent
```

```python
data = pd.merge(students , marks , on="student_id")
data
```

```python
data = pd.merge(students , marks , on="student_id" , how="inner")
data
```

```python
data = pd.merge(students , marks , on="student_id" , how="left")
data
```

```python
data = pd.merge(students , marks , on="student_id" , how="right")
data
```

```python
data = pd.merge(students , marks , on="student_id" , how="outer")
data
```

```python
data = pd.concat([students , ExtraStudent] , ignore_index=True)
data
```
